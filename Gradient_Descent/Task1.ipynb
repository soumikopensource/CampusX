# Optimization of error in regression model using cost function

```python
#formula of the cost function= J(@) = @1+@2x2+@3x3+@4x4+..........
```

```python
def gradient_descent(x,y,learning_rate,n,max_iterations):
    """
    we have multiple features so we have to calculate theta for each feature 
    and aggregate that to the main equation
    
    """

    slope=1
    intercept=1
    iteration= 0
    y_updated=slope * x + intercept
    error= 0
    while(True):
        error=sum((y_updated - y) **2) / n
        
        print("error after {} iteration: {}\n".format(iteration,error))
        
        
        slope_diff=-(2/len(x_train))*sum((y_train -y_updated)*x_train.reshape(400))
        intercept_diff=-(2/len(x_train))*sum(y_train -y_updated)
        slope_updated=slope - learning_rate * slope_diff
        intercept_updated=intercept - learning_rate * intercept_diff
        slope=slope_updated
        intercept=intercept_updated
        
        y_updated=slope * x + intercept
        
        error= sum((y - y_updated) ** 2) / n
        
        
        
        #print(intercept)
        #print(slope)
        iteration+=1
        if(iteration> max_iterations):
            break
    
        
    
    
    
    
    
```
